```python
#using nrows method we can assign the number of the rows we want to get from the data
df = pd.read_csv('https://raw.githubusercontent.com/ShohruhShokulov/Probability-and-Statistics/main/Charts/responses.csv',index_col = 0,nrows = 10)
df
```
