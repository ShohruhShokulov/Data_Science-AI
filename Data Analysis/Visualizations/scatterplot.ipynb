```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('https://github.com/anvarnarz/praktikum_datasets/raw/main/merc.csv')
df.head()
```

```python
#scatterplot
df = df[df.year>2000]
plt.figure(figsize = (10,6))
sns.scatterplot(data = df, x = 'year', y = 'price')
plt.show()
```

```python
#showing the correlation between two variables
plt.figure(figsize = (10,6))
sns.regplot(data = df, x = 'year', y = 'price')
plt.show()
```

```python
df = df[df.year>2000]
plt.figure(figsize = (10,6))
sns.scatterplot(data = df, x = 'mileage', y = 'price')
plt.show()
```

```python
# plot with more variables
plt.figure(figsize = (16,8))
sns.scatterplot(data = df, x = 'mileage', y = 'price', hue = 'transmission', s = 100) # s is for the size of the points where hue - adding new variable
plt.show()
```

```python
# plot with more variables
plt.figure(figsize = (16,8))
sns.scatterplot(data = df, x = 'mileage', y = 'price', hue = 'transmission', style = 'fuelType', s = 100) # style is for another variable
plt.show()
```

```python
plt.figure(figsize = (16,8))
sns.scatterplot(data = df, x = 'mileage', y = 'price', hue = 'fuelType', s = 100) # s is for the size of the points where hue - adding new variable
plt.show()
```
